```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# import our LA_Crimes into pandas 
LA_Crimes_df = pd.read_csv("Resources/2016_LA_Crimes.csv")
#LA_Crimes_df.head()
```

```python
# import our LA_Crimes_List into pandas 
LA_Crimes_list_df = pd.read_csv("Resources/LA_Crimes_List.csv")
```

```python
# Merge the 2 dataframes using left join

LA_Crime_Complete_df = pd.merge(LA_Crimes_df, LA_Crimes_list_df, how="left", on=["Description", "Description"])
LA_Crime_Complete_df.head(10000)
```

```python
LA_Crimes_list_df.groupby("Type").count()
```

```python
#Define each type
Theft_type_df = LA_Crime_Complete_df[LA_Crime_Complete_df["Type"] == "Theft"]
Theft_avg =np.int(Theft_type_df.groupby(["Type"]).count()["Count"])

Assault_type_df = LA_Crime_Complete_df[LA_Crime_Complete_df["Type"] == "Assault"]
Assault_avg =np.int(Assault_type_df.groupby(["Type"]).count()["Count"])

Destruction_of_Property_type_df = LA_Crime_Complete_df[LA_Crime_Complete_df["Type"] == "Destruction of Property"]
Destruction_of_Property_avg =np.int(Destruction_of_Property_type_df.groupby(["Type"]).count()["Count"])

Disruption_type_df = LA_Crime_Complete_df[LA_Crime_Complete_df["Type"] == "Disruption"]
Disruption_avg =np.int(Disruption_type_df.groupby(["Type"]).count()["Count"])

Miscellaneous_type_df = LA_Crime_Complete_df[LA_Crime_Complete_df["Type"] == "Miscellaneus"]
Miscellaneous_avg =np.int(Miscellaneous_type_df.groupby(["Type"]).count()["Count"])

Fraud_type_df = LA_Crime_Complete_df[LA_Crime_Complete_df["Type"] == "Fraud"]
Fraud_avg =np.int(Fraud_type_df.groupby(["Type"]).count()["Count"])

Crimes_against_children_type_df = LA_Crime_Complete_df[LA_Crime_Complete_df["Type"] == "Crimes against children"]
Crimes_against_children_avg =np.int(Crimes_against_children_type_df.groupby(["Type"]).count()["Count"])

Sexual_type_df = LA_Crime_Complete_df[LA_Crime_Complete_df["Type"] == "Sexual"]
Sexual_avg =np.int(Sexual_type_df.groupby(["Type"]).count()["Count"])

Legal_type_df = LA_Crime_Complete_df[LA_Crime_Complete_df["Type"] == "Legal"]
Legal_avg =np.int(Legal_type_df.groupby(["Type"]).count()["Count"])

Human_Trafficking_type_df = LA_Crime_Complete_df[LA_Crime_Complete_df["Type"] == "Human Trafficking"]
Human_Trafficking_avg =np.int(Human_Trafficking_type_df.groupby(["Type"]).count()["Count"])



print(Theft_avg)
print(Assault_avg)
print(Destruction_of_Property_avg)
print(Disruption_avg)
print(Miscellaneous_avg)
print(Fraud_avg)
print(Crimes_against_children_avg)
print(Sexual_avg)
print(Legal_avg)
print(Human_Trafficking_avg)
```

```python
# create a dataframe 

LA_Crime_Complete_pd = pd.DataFrame({"Theft":[Theft_avg],
                                     "Assault":Assault_avg,
                                     "Destruction of Property":Destruction_of_Property_avg,
                                     "Disruption":Disruption_avg,
                                     "Miscellaneous":Miscellaneous_avg,
                                     "Fraud":Fraud_avg,
                                     "Crimes against children":Crimes_against_children_avg,
                                     "Sexual":Sexual_avg,
                                     "Legal":Legal_avg,
                                     "Human Trafficking":Human_Trafficking_avg})

LA_Crime_Complete_pd
```
